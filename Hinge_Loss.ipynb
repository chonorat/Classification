```
import sys
import random

####################################
##############Dot Product
#################################
def dot_prod(W_Vector,F_Vector):
    dp=0
    for i in range(0,len(W_Vector)):
        dp=dp+(W_Vector[i]*(F_Vector[i]))
    return(dp)
####################
###Read Data############
######################
datafile='/content/ionosphere.data'
f=open(datafile)
data=[]
i=0
l=f.readline()
while(l!=''):
        a=l.split()
        l2=[]
        for j in range(0,len(a),1):
                l2.append(float(a[j]))
        data.append(l2)
        l=f.readline()
####
###Add column of ones to data
######
rows=len(data)
for i in range(0,rows):
        data[i].append(1)
f.close()
cols=len(data[0])
#################
##################Read Labels
###################
labelfile='/content/ionosphere_trailabels.data'
f=open(labelfile)
trainlabels={}
n=[]
n.append(0)
n.append(0)
l=f.readline()
while(l !=''):
        a=l.split()
        if int(a[0])==1 or int(a[0])==0:
                trainlabels[int(a[1])]=int(a[0])
                n[int(a[0])]+=1
        l=f.readline()
################
##Initiate Dellf
################

dellf=[]
for j in range(0,cols,1):
        dellf.append(0)
		
##################
##set class labels
##################
for i in range(0,rows-1):
        if(trainlabels.get(i)!=None and trainlabels[i]==0):
                trainlabels[i]=-1
				
############################
##########Initialize weights
############################
w=[]
for k in range(0,cols,1):
        w.append(0)
for k in range(0,cols,1):
        w[k]=.02 * random.uniform(0,1) - .01
		
############################
##########Initialize weights
############################
w=[]
for k in range(0,cols,1):
        w.append(0)
for k in range(0,cols,1):
        w[k]=.02 * random.uniform(0,1) - .01
		
#############################
#############Gradient Descent
##############################
eta=.001
error_diff=1
error=0
dotprod=0
while error_diff>.001:
#for p in range(0,2):
        ###Compute dellf
        for i in range(0,rows):
                if(trainlabels.get(i)!=None):
                        dotprod=dot_prod(w,data[i])
                        for j in range(0,cols):
                                #print(trainlabels[i]*dotprod)
                                if ((trainlabels[i]*dotprod)**2)**.5<1:
                                        dellf[j]=dellf[j]-(trainlabels[i]*data[i][j])
        ####update
        for j in range(0,cols):
                w[j]=w[j]-(eta*dellf[j])
                dellf[j]=0
        ###compute error and error_difference
        error_previous=error
        error=0
        for i in range(0,rows):
                if(trainlabels.get(i)!=None):
                        if ((trainlabels[i]*dot_prod(w,data[i]))**2)**.5<1:
                                error=error+1-(trainlabels[i]*dot_prod(w,data[i]))
        error_diff=((error_previous-error)**2)**.5
        print("ERROR= ",error_diff,"\n")

#Calculate norm
normw=0
for i in range(0,cols-1):
        normw=normw+w[i]**2
normw=(normw)**.5
#w[-1]=((w[-1]**2)**.5)

###############
####Predictions
###############
for i in range(0,rows-1):
        if (trainlabels.get(i)==None):
                dp=dot_prod(w,data[i])
                if (dp)>0:
                        print("Prediction ",i," = 1","\n")
                else:
                        print("Prediction ",i," = 0","\n")
print("DISTANCE TO ORIGIN= ",w[-1]/normw,"\n")
print("WEIGHTS= ",w)
print(normw)
```
