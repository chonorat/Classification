```
import sys
import numpy as np
import random

##Define nmber of clusters
k=2

##Read Data
datafile='/content/ionosphere.data'
f=open(datafile)
l=f.readline()

data=[]
while(l!=''):
        a=l.split()
        l2=[]
        for j in range(0,len(a),1):
                l2.append(float(a[j]))
        data.append(l2)
        l=f.readline()
rows=len(data)
cols=len(data[0])
f.close()

##Dot Product
def dot_prod(W_Vector,F_Vector):
        dp=0
        for i in range(0,len(W_Vector)):
                dp=dp+(W_Vector[i]*(F_Vector[i]))
        return(dp)

def create_clust_means(data,rand,k):
        m=[]
        for i in range(0,k):
                to_append=[]
                to_append.append(i)
                for j in range(0,cols):
                        if rand==1:
                                to_append.append( random.choice(data[j]))
                        else: to_append.append(0)
                m.append(to_append)
        return m
		
def classify(m,data,k):
        pred=[]
        for i in range(0,rows):
                to_append=[]
                d_small=1000000
                for m_index in range(0,k):
                        d=0
                        for j in range(0,cols):
                                d=d+(m[m_index][j]-data[i][j])**2
                        if(d<d_small):
                                classs=m_index
                                d_small=d
                to_append.append(classs)
                #print(1,i)
                to_append.append(i)
                pred.append(to_append)
        return pred

def calculate_means(pred,data,cols,k):
        m=create_clust_means(data,0,k)
        for classs in range(0,k):
                n=0
                for i in range(0,len(pred)):
                        index=pred[i][1]
                        if( pred[i][0]==classs):
                                for j in range(0,cols):
                                        m[classs][j]=m[classs][j]+data[index][j]
                                        n=n+1
                for colss in range(0,cols):
                        if n>0:
                                m[classs][colss]=m[classs][colss]/n
        return m
		
def err(m,data,pred,k):
        error=0
        for classs in range(0,k):
                for i in range(0,len(pred)):
                        index=pred[i][1]
                        if (pred[i][0]==classs):
                                for j in range(0,cols):
                                        error=error+(m[classs][j]-data[index][j])**2
        return error
		
##Main method
m=create_clust_means(data,1,k)
for i in range(0,1000):
        #1. Classify training data
        pred=classify(m,data,k)
        #2. Calculate Loss
        error=err(m,data,pred,k)
        #print(error)
        #3. Re-calculate means according to train predictions
        m=calculate_means(pred,data,cols,k)

def classify_test(m,data,k):
        pred=[]
        for i in range(0,rows):
                to_append=[]
                d_small=1000000
                for m_index in range(0,k):
                        d=0
                        for j in range(0,cols):
                                d=d+(m[m_index][j]-data[i][j])**2
                        if(d<d_small):
                                classs=m_index
                                d_small=d
                print(classs,i)
##Pint predicted labels
pred=classify_test(m,data,k)

```
